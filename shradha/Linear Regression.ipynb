# Linear Regression

```python
##  splitting data into x as tv and y as sales then splitting x as training and testing and same in case of y

X = df[["TV"]]
y = df[["sales"]]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
##0.3 means 70% training and 30% testing should be performed & random_states means for same data checking for our purpose 
```

```python
## an object of linear regression lr
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.intercept_ ## theta 0
```

```python
## theta 1 slope 1 unit change in tv 0.048  change in sales if 0 is tv then the value will be 6.9 i.e. intercept_
lr.coef_
```

```python
y_pred = lr.predict(X_test)# y_pred is the predicted o/p
```

```python
y_pred
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.figure()
plt.scatter(df["TV"],df["sales"])
plt.plot(X_test,y_pred)
plt.show()
```

# Gradient Descent

```python
def gradient_descent(x,y):
    n = len(x)
    learning_rate = 0.01 ## step size
    intercept = slope = 0
    iterations = 500## no. of steps
    
    for i in range(iterations):
        
        y_pred = x*slope + intercept
        
        mse = mean_squared_error(y_pred,y)
        
        md = (2/n)*sum(x*(y-y_pred)) 
        bd = (2/n)*sum(y-y_pred) 

        slope = slope + (learning_rate * md)
        intercept = intercept + (learning_rate * bd)
        
        print("slope: ",slope,"\nintercept: ",intercept,"\nmse: ",mse,"iteration: ",i)
```

```python
x = np.array([1,2,3,4])
y = np.array([12,15,16,18])
```

```python
gradient_descent(x,y)
```

```python
def gradient_descent(x,y):
    n = len(x)
    learning_rate = 0.000001 ## step size
    intercept = slope = 0
    iterations = 150## no. of steps
    
    for i in range(iterations):
        
        y_pred = x*slope + intercept
        
        mse = mean_squared_error(y_pred,y)
        
        md = (2/n)*sum(x*(y-y_pred)) 
        bd = (2/n)*sum(y-y_pred) 

        slope = slope + (learning_rate * md)
        intercept = intercept + (learning_rate * bd)
        
        print("slope: ",slope,"\nintercept: ",intercept,"\nmse: ",mse,"\niteration: ",i,"\n............")
        
```

```python
## u can keep on increasing iteration or reducing learning_rate and decresing mse value can give a line close to best fit line below 1 value of mse is considered as good 
x= df["TV"]
y= df["sales"]
```

```python
gradient_descent(x,y)
```

```python
millis_before = int(round(time.time() * 1000))
gradient_descent(x,y)
millis_after = int(round(time.time() * 1000))
print(millis_after - millis_before)
```

```python
## visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [11,13,18,19,21,24]
```

```python
plt.plot(x,y)
```

```python
## to remove the written details then show is used
plt.plot(x,y)
plt.show()
```

```python
## 0.1 is step size
x = np.arange(0,10,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
## figure should be written before plotting
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
## limit can be even written without tuple also
plt.xlim((200,2000))
plt.ylim((5,50))

plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)

plt.xticks(rotation=90)
plt.yticks(rotation=45)

plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
x = np.arange(1,500)

y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.figure(figsize=(7,7))
plt.plot(x,y1,color="blue")
plt.plot(x,y2,color="red")
plt.legend(labels=["sqrt","log"]) # even if one element is there then also pass in list 
plt.show()
## log is preferably used first to reduce skewness after some pt of time log starts becoming constant and then sqrt is used but it takes longer time
```

## ### HISTOGRAMS 

```python
##Histogram is a plot of continuous data which can be measured
df = pd.read_csv("cars.csv")
```

```python
df.info()
```

```python
df.describe() #considers only numerical columns
```

```python
plt.figure(figsize=(7,4))
plt.hist(df["width"],bins=5)## shows distribution range of x axis is called bins


plt.title("Width Distribution")
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.hist(df["highway-mpg"],bins=5)## shows distribution range of x axis is called bins


plt.title("Highway Distribution")
plt.xlabel("Hway")
plt.ylabel("Frequency")
plt.show()
```

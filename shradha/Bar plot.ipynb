##### BAR PLOT


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cars.csv")
```

```python
df["make"].value_counts().plot(kind="bar")
plt.show()
```

```python
df["make"].value_counts().plot(kind="barh")
plt.show()
```

```python
df["make"].value_counts().plot(kind="pie")
plt.show()
```

```python
df["body-style"].value_counts().plot(kind="bar")
plt.show()
```

```python
# autopct means pie along with the %age values & explore the separated one
df["body-style"].value_counts().plot(kind="pie",autopct="%1.1f%%",explode=(0,0,0.3,0,0))
plt.show()
```

```python
## scatter plots only will be performed on numerical columns
df.corr()
```

```python
## positive correlation 
plt.scatter(df["engine-size"],df["price"])
plt.show()
```

```python
## negative correlation
plt.scatter(df["highway-mpg"],df["price"],color="red")
plt.show()
```

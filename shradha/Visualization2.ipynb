```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv")
```

```python
df
```

```python
x = np.arange(20)
y = np.sqrt(x)
```

```python
## to plot a line in seaborn
sns.lineplot(x,y)
plt.show()
```

```python
## to plot a line in matplotlib
plt.plot(x,y)
plt.show()
```

```python
## to plot a histogram in seaborn
sns.distplot(df["width"])
plt.show()
```

```python
## to see categorical data in bar graph for seaborn first argument is data frame to be used and x is column
sns.countplot(data=df, x="body-style")
plt.show()
```

```python
sns.countplot(data=df, x="make")
plt.show()
```

```python
## to make it more presentable then we can shift to y axis
sns.countplot(data=df, y="make")
plt.show()
```

```python
sns.countplot(data=df, x="body-style", hue="fuel-type")
plt.show()
```

```python
sns.countplot(data=df, x="drive-wheels",hue="fuel-type")
plt.show()
```

# HEATMAP


```python
corr = df.corr()
```

```python
sns.heatmap(corr)## all columns correlation dark colors -vely correlated and faint colors +vely correlated
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap="viridis")## values 
plt.show()
```

# Scatter plot

```python
sns.scatterplot(data=df ,x="width", y="price" , hue="body-style")
plt.show()
```

# Box plot

```python
## box plot can detect the outliers the dotd are outliers , brand wise outliers should be removed not by price wise and high variance means range is high so mercedus-benz,porsche and low var is renault ,etc 
plt.figure(figsize=(15,15))
sns.boxplot(data=df,x="price", y="make")
plt.show()
```

# SUB PLOTS

```python
# to create subplot fig,axes it returns a tuple of two values the inner plots are matplotlib plots because only plot is used nt line plot in seaborn
fig, axes =plt.subplots(2,2,figsize=(6,6))
axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].scatter(x,y)
axes[1,1].plot(x,y)
plt.show()
```

```python
## here inner plots are seaborn plots as sns is used 
fig, axes =plt.subplots(2,2,figsize=(6,6))
sns.boxplot(data=df,x="price", y="body-style", ax=axes[0,0])
sns.scatterplot(data=df ,x="width", y="price" , hue="body-style", ax=axes[0,1])
sns.countplot(data=df, x="drive-wheels",hue="fuel-type",ax=axes[1,0])
sns.heatmap(corr, annot=True, cmap="viridis",ax=axes[1,1])
plt.show()
```
